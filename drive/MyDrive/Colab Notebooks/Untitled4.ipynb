```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical

```

```python
#LOAD CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

```python
input_shape = (32, 32, 3)
num_classes = 10
alexnet_model = models_Sequential()

#layer 1:conv layer

alexnet_model.add(layers.Conv2D(96, (11, 11), strides=4, padding='same', input_shape=input_shape))
alexnet_model.add(layers.BatchNormalization())
alexnet_model.add(layers.Activation('relu'))
alexnet_model.add(MaxPooling2D(pool_size=(2, 2), strides=(1, 1)))

#layer 2 conv layer

alexnet_model.add(Conv2D(256, (5, 5), padding='same'))
alexnet_model.add(layers.BatchNormalization())
alexnet_model.add(layers.Activation('relu'))
alexnet_model.add(MaxPooling2D(pool_size=(2, 2), strides=(1,1)))

#layer 3 conv layer

alexnet_model.add(Conv2D(384, (3, 3), padding='same'))

#layer 4 conv layer


alexnet_model.add(Conv2D(384, (3, 3), padding='same'))

#layer 5 convo layer

alexnet_model.add(Conv2D(256, (5, 5), padding='same'))
alexnet_model.add(layers.Activation('relu'))
alexnet_model.add(MaxPooling2D(pool_size=(2, 2), strides=(1, 1)))

#flatten the dataset for fully connected layers

alexnet_model.add(Flatten())


alexnet_model.add(Dense(4096))
alexnet_model.add(layers.Activation("relu"))
alexnet_model.add(layers.Dropout(0.5))

alexnet_model.add(Dense(4096))
alexnet_model.add(layers.Activation("relu"))
alexnet_model.add(layers.Dropout(0.5))

alexnet_model.add(Dense(num_classes))
alexnet_model.add(layers.Activation("softmax"))


alexnet_model.compile(optimizer=Adam(learning_rate=0.001), loss="categorical_crossentropy", metrics=["accuracy"])


alexnet_model.summary()


tf.keras.utils.plot_model(alexnet_model, show_layer_names=True, show_shapes=True,show_dtype=False)

alexnet_model.add(layers.Activation("relu"))
alexnet_model.add(layers.MaxPooling2D(pool_size=(2,2), strides=(1,1)))


alexnet_model.add(Flatten())


alexnet_model.add(Dense(4096))
alexnet_model.add(layers.Activation("relu"))
alexnet_model.add(layers.Dropout(0.5))

alexnet_model.add(Dense(4096))
alexnet_model.add(layers.Activation("relu"))
alexnet_model.add(layers.Dropout(0.5))

alexnet_model.add(Dense(num_classes))
alexnet_model.add(layers.Activation("softmax"))


alexnet_model.compile(optimizer=Adam(learning_rate=0.001), loss="categorical_crossentropy", metrics=["accuracy"])


alexnet_model.summary()


tf.keras.utils.plot_model(alexnet_model, show_layer_names=True, show_shapes=True,show_dtype=False)

```
